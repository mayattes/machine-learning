```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([ 1, 2, 3, 4, 5], [1, 3, 2, 4, 3])
plt.title("Prix de stock")
plt.xlabel("Temps(s)")
plt.ylabel("Prix($)")
plt.show()
```

```python
t = [1,2,3,4,5]
prix_microsoft = [1,2,3,4,8]
prix_apple = [2,4,4,3,6]
plt.plot(t, prix_microsoft, label = "microsoft")
plt.plot(t, prix_apple, label = "apple")
plt.legend()
plt.title("Prix de stock")
plt.xlabel("Temps (s)")
plt.ylabel("Prix ($)")
plt.show()
```

```python
b = [1,2,3,4,5,6]
prix_jupe = [1,3,2,4,6,5]
prix_pantalon = [2,4,3,6,5,4]
plt.plot(b, prix_jupe, label = "Prix des jupes")
plt.plot(b, prix_pantalon, label = "Prix des pantalon")
plt.title("Prix des stock")
plt.xlabel("Prix des pantalons")
plt.ylabel("prix des jupes")
plt.legend()
plt.show()
```
