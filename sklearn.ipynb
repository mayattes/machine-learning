```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('iris.csv')
dataset.head()
```

```python
plt.plot(dataset['petal_length'], dataset['petal_width'], 'o')
plt.xlabel("petal_length")
plt.ylabel("petal_width")
plt.show()
```

```python
from sklearn import linear_model

X = dataset['petal_length'].values
Y = dataset['petal_width'].values

#print(X)
X = X.reshape(-1, 1)
Y = Y.reshape(-1, 1)
#print(X)

modele = linear_model.LinearRegression()
modele.fit(X, Y)

Y_pred = modele.predict(X)
#print(modele.coef_,modele.intercept_)
```

```python
plt.plot(X, Y, 'o',label='dataset')
plt.plot(X, Y_pred, label = 'prdictions')
plt.xlabel('Longeur de petal')
plt.ylabel('Largeur de petal')
plt.legend()
plt.show()
```

```python
y_pred = modele.predict([[4], [3]])
y_pred
```

# erreur quadratique

```python
from sklearn.metrics import mean_squared_error

erreur = mean_squared_error(Y_pred, Y)
erreur
```
