```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
w = 5
b = 0
x = np.linspace(0, 10, 11)

y = b + w*x

print(x)
print(y)
```

```python
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
axes = plt.gca()
axes.set_xlim([0, 10])
axes.set_ylim([0, 50])
plt.show()
```
